## Universal functions (numpy ufunc)

To round the data we first used the universal function ``np.floor``. A universal function is a function that is applied element-wise on an array (or data frame), yileding new array (or dataframe).

We can combine the numpy ufuncs to create new ufuncs that we might be interested in.

In this example, we will create a completely useless function that converts the temperature from Celsius to Kelvin before taking the logarithm.

```python
def log_kelvin(temperature):
    return np.log(temperature + 273)
```

```python
weather['log_kelvin_lt'] = log_kelvin(weather['lt'])
```

```python
weather['log_kelvin_lt']
```

## Advanced indexing with groupby

We want to look whether the summary statistics for the dataset vary across the different seasons. To do this, we start by adding a month column to the DataFrame.

```python
weather['month'] = weather.index.month
```

```python
weather['month']
```

```python
weather.groupby(weather['month'])
```

When we group by ``weather['month']``, we separate the weather data frame into twelve new dataframes. Each row in ``weather``, whose corresponding row in ``weather['month']`` is equal to 1 is placed in the first data frame. Similarly,  the rows in ``weather`` that match up with the rows where ``weather['month']`` is equal to two, are placed in the secon data frame.

```python
weather.groupby(weather['month']).mean()
```

We can also group by multiple variables, let us group by year and month.

```python
weather['year'] = weather.index.year
```

```python
weather.groupby([weather['year'], weather['month']]).mean()
```

### This is not the best way to group data if you are grouping by time.

We want to use the pandas Grouper object

```python
weather.groupby(pd.Grouper(freq='M')).mean()
```

We can store the grouped dataset in a variable

```python
weekly_weather_mean = weather.groupby(pd.Grouper(freq='W')).mean()
```

```python
weekly_weather_mean
```

```python
weather.groupby(weather.index.week).mean()
```

## Statistics in Python

We can use several tools for statistics in Python, the two most common are StatsModels and scipy.stats. We will use the former to perform a simple ANOVA analysis

### One-way ANOVA
We start by performing an ANOVA to test whether the month variable is statistically significant.

```python
from statsmodels.formula.api import ols
```

```python
model = ols('lt ~ C(month, Treatment)', data=weather).fit()
```

```python
model.summary()
```

```python
model = ols('lt ~ 0 + C(month, Treatment)', data=weather).fit()
model.summary()
```

Converting this to a one-way ANOVA table.

```python
from statsmodels.stats.anova import anova_lm
```

```python
anova_lm(model)
```
